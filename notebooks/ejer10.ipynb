```python

import sys
sys.path.append("..")  # Esto asume que estás dentro de /ejercicios y querés ir a la raíz
from src.Juego import calcular_puntos
#Lista de rondas de cada jugador 

rounds = [
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
 'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
 'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
 'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
 'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
 }
 ]
 #round es una lista donde cada elemento es un diccionario con los jugadores y sus respectivas estadisticas, teniendo 3 valores claves

 #diccionario para almacenar las estadisticas totales de cada jugador
jugadores_puntos = {}

#contador de MVP por jugador
mpv = {}

#¿como llenaremos este diccionario?, bueno recorremos cada ronda y cada jugador, y sumamos las estadisticas de cada jugador
#recorremos cada ronda 
for numero_ronda, ronda in enumerate(rounds, start=1):
    print(f"\nRanking ronda {numero_ronda}:")
    mvp_ronda = None
    max_puntos = float('-inf')
    puntos_ronda_jugadores = {}
    #recorre el diccionario ronda y obtiene cada jugador junto con sus estadísticas.
    for jugador , stats in ronda.items():
        puntos_ronda = calcular_puntos(stats["kills"], stats["assists"], stats["deaths"])#llama a la funcion calcular puntos y le pasa los valores de cada jugador

         #Inicializamos el jugador si no existe en el diccionario 
        if jugador not in jugadores_puntos:
            jugadores_puntos[jugador] = {"kills": 0, "assists": 0, "deaths": 0, "puntos": 0}
            mpv[jugador] = 0
        #Actualizamos estadisticas del jugador
        jugadores_puntos[jugador]["kills"] += stats["kills"]
        jugadores_puntos[jugador]["assists"] += stats["assists"]
        jugadores_puntos[jugador]["deaths"] += 1 if stats["deaths"] else 0
        jugadores_puntos[jugador]["puntos"] += puntos_ronda

        # Guardamos los puntos de esta ronda para determinar al MVP
        puntos_ronda_jugadores[jugador] = puntos_ronda

        # Comprobamos si este jugador es el nuevo MVP
        if puntos_ronda > max_puntos:
            max_puntos = puntos_ronda
            mvp_ronda = jugador

# Sumamos un MVP al mejor jugador de la ronda
    if mvp_ronda not in mpv:
        mpv[mvp_ronda] = 0
    mpv[mvp_ronda] += 1

    # Mostramos el ranking parcial ordenado por puntos
    ranking_parcial = sorted(jugadores_puntos.items(), key=lambda x: x[1]["puntos"], reverse=True)
    for jugador, stats in ranking_parcial:
        print(f"{jugador:10} | Kills: {stats['kills']} | Asists: {stats['assists']} | Muertes: {stats['deaths']} | Puntos: {stats['puntos']} | MVPs: {mpv[jugador]}")

# Ranking final # "Mostramos un título para que el jugador sepa que ya terminó el torneo y viene el resultado acumulado."
#"Ordenamos a todos los jugadores según los puntos totales que acumularon. Usamos la función sorted() y como clave le pasamos los puntos, para que el jugador con más puntos quede primero. El reverse=True hace que el orden sea de mayor a menor."
print("\n Ranking Final:")
ranking_final = sorted(jugadores_puntos.items(), key=lambda x: x[1]["puntos"], reverse=True)
#"Ahora recorremos ese ranking ordenado y mostramos por pantalla las estadísticas acumuladas de cada jugador: puntos totales, kills, assists, muertes y la cantidad de veces que fue MVP."
for jugador, stats in ranking_final:
    print(f"{jugador:10} | Total Puntos: {stats['puntos']} | Kills: {stats['kills']} | Assists: {stats['assists']} | Muertes: {stats['deaths']} | MVPs: {mpv[jugador]}")

# Mostramos MVP del torneo "Finalmente, usamos max() para encontrar quién fue el jugador con más MVPs y lo mostramos como el MVP del torneo."
mvp_torneo = max(mpv.items(), key=lambda x: x[1])
print(f"\n MVP del Torneo: {mvp_torneo[0]} con {mvp_torneo[1]} MVPs")
```
